```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_path = os.path.join(os.getcwd(), 'ynap_data')
print(data_path)
```

```python
df_tot = pd.read_csv(os.path.join(data_path, 'df_tot.csv'))
```

```python
df_tot.head()
```

```python
df_tot.drop('customer_id', axis=1).corr()
```

```python
fig, ax = plt.subplots(figsize=(12,10))
g = sns.heatmap(df_tot.drop('customer_id', axis=1).corr(), annot=True,
                fmt=".2f", cmap='viridis', ax=ax)
```

```python
g = sns.pairplot(df_tot.drop('customer_id', axis=1), hue='lapsed_next_period', size=1.2, diag_kind='kde',
                 diag_kws=dict(shade=True), plot_kws=dict(s=10))
g.set(xticklabels=[])
```

```python
# g = sns.FacetGrid(data=df_tot, hue='lapsed_next_period')
g = sns.kdeplot(df_tot['quote_var2'][df_tot['lapsed_next_period'] == 0], color="Red", shade = True)
g = sns.kdeplot(df_tot['quote_var2'][df_tot['lapsed_next_period'] == 1], ax =g, color="Blue", shade= True)
# g.despine(left=True)
```
